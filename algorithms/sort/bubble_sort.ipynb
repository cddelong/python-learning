# Bubble Sort
Runtime: O(n^2) average and worst case.

Memory: O(1)
### Algorithm:
In bubble sort we start at the beginning of the array and swap the first two elements if the first is greater than the second.
Then, we go to the next pair and so on, continuously making sweeps of the array until it is sorted.
In doing so, the smaller items slowly "bubble" up to the beginning of the list.

Lower to the left; Higher to the right
### In pseudocode
* Set a "swap" counter to a non-zero value
* Repeat until the swap counter is 0:
    * reset swap counter to 0
    * Look at each adjacent pair
        * if two adjacent elements are not in order, swap them and ad one to the swap counter

```python
from random import random
n = 8500
test_array = [
     int(random()*100) for _ in range(n)
]
```

```python
from copy import copy

def bubble_sort(array: list):
    iterations = 0;    swap = 1
    while swap > 0:
        swap = 0
        for i in range(len(array)-1):
            iterations += 1
            if array[i] > array[i+1]:
                # copy is used here so that I can overwrite the value at array[i] without loosing it
                arr_i = copy(array[i])
                swap += 1
                array[i] = array[i+1]
                array[i+1] = arr_i
    return f'iterations: {iterations}', array
```

```python
res = bubble_sort(test_array)
print(res)
```
